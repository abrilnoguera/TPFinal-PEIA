```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.2f' % x)
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import t as t_dist  # Utilizamos solo para comparar el p-valor y el valor crítico
import statsmodels.api as sm
```

```python
# Cargamos las dos hojas como DataFrames
df_santa_ana = pd.read_excel('Datos.xlsx', sheet_name='Santa Ana')
df_santa_ana['Tienda'] = 'Santa Ana'
df_la_floresta = pd.read_excel('Datos.xlsx', sheet_name='La Floresta')
df_la_floresta['Tienda'] = 'La Floresta'

df = pd.concat([df_santa_ana, df_la_floresta], ignore_index=True)
```

```python
# Procesamiento de la variable 'Fecha'
df['Fecha'] = pd.to_datetime(df['Fecha'])

df['Mes'] = df['Fecha'].dt.month
df['Dia'] = df['Fecha'].dt.weekday
df['DiaSemana'] = df['Fecha'].dt.dayofweek
dias = {0: 'Lunes', 1: 'Martes', 2: 'Miércoles', 3: 'Jueves', 4: 'Viernes', 5: 'Sábado', 6: 'Domingo'}
df['DiaNombre'] = df['DiaSemana'].map(dias)
```

```python
# Agrupamos por Mes y Tienda, sumando ventas
tabla_ventas = df.groupby(['Mes', 'Tienda'])['Ventas'].sum().unstack()

# Agregamos la columna 'Total' sumando Santa Ana + La Floresta
tabla_ventas['Total'] = tabla_ventas.sum(axis=1)

# Mostramos la tabla final
print(tabla_ventas)
```

```python
# Graficamos la tabla de ventas
colores = ['#4F81BD', '#9BBB59']

tabla_ventas[['Santa Ana', 'La Floresta']].plot(
    kind='bar',
    stacked=True,
    figsize=(12,7),
    color=colores
)

# Agregamos la línea de promedio
promedio_total = tabla_ventas['Total'].mean()
plt.axhline(y=promedio_total, color='crimson', linestyle='--', label=f'Promedio Total: {promedio_total:.2f}')

# Mejoramos el gráfico
plt.title("Distribución Empirica de Ventas por Mes")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.xticks(rotation=0)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostramos
plt.show()
```

A partir del análisis de las ventas mensuales, se observó que los **meses de menores ventas** fueron **enero** y **febrero**, con ventas totales aproximadas de **1.096.164 dólares** y **1.092.859 dólares** respectivamente, valores muy similares y claramente por debajo del promedio anual de **1.305.432 dólares**.

Por otro lado, se identificaron dos momentos de **ventas máximas**: el primero en **marzo**, con aproximadamente **1.386.045 dólares** en ventas, y el segundo, aún más elevado, en **agosto**, alcanzando un total de **1.438.324 dólares**.

En base a estos resultados, se recomienda a Don Francisco considerar **enero o febrero** como los meses más adecuados para tomar vacaciones, dado que representan los períodos de menor actividad comercial. Asimismo, **marzo** y especialmente **agosto** surgen como los momentos más convenientes para **realizar inversiones importantes**, coincidiendo con los meses de mayores ingresos.

```python
# Agrupamos ventas por tienda y día de la semana
ventas_por_dia_tienda = df.groupby(['DiaNombre', 'Tienda'])['Ventas'].sum().unstack()

# Ordenamos los días correctamente
ventas_por_dia_tienda = ventas_por_dia_tienda.reindex(['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'])

# Agregamos la columna 'Total' sumando Santa Ana + La Floresta
ventas_por_dia_tienda['Total'] = ventas_por_dia_tienda.sum(axis=1)

# Mostramos la tabla final
print(ventas_por_dia_tienda)
```

```python
# Gráfico de comparación
ventas_por_dia_tienda[['Santa Ana', 'La Floresta']].plot(
    kind='bar',
    stacked=True,
    figsize=(12,7),
    color=colores
)

# Agregamos la línea de promedio
promedio_total = ventas_por_dia_tienda['Total'].mean()
plt.axhline(y=promedio_total, color='crimson', linestyle='--', label=f'Promedio Total: {promedio_total:.2f}')

# Mejoramos el gráfico
plt.title("Distribución Empirica de Ventas por Mes")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.xticks(rotation=0)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostramos
plt.show()
```

Del análisis surge que el **miércoles** y el **jueves** son los días de **mayor volumen de ventas**, con valores superiores al promedio semanal de aproximadamente **2.237.884 dólares**. Estos días presentan una demanda más alta, por lo que se recomienda reforzar el personal para asegurar un buen nivel de atención al cliente.

En contraste, el **domingo** resulta ser el día de **menor volumen de ventas**, seguido por el **lunes** y el **sábado**. Estos días presentan valores de venta considerablemente más bajos en comparación con el resto de la semana. Por lo tanto, Don Francisco podría considerar reducir el personal durante domingos y lunes para optimizar los costos operativos.

En general, se observa que la actividad comercial se intensifica hacia mediados de semana, con un descenso hacia el fin de semana, lo que sugiere que la planificación de horarios y la asignación de recursos humanos debe estar alineada con este patrón de comportamiento.

Si bien la diferencia no es extremadamente significativa, muestra que "La Floresta" se mantiene competitiva e incluso supera ligeramente a "Santa Ana" en desempeño comercial.

```python
for mes in meses_santa:
    datos = santa_ana_df[santa_ana_df['Mes'] == mes]['Ventas']
    x = np.sort(datos)
    y = np.arange(1, len(x)+1) / len(x)    

    # Intervalos de Confianza Empíricos
    n = len(datos)
    datos_ordenados = np.sort(datos)
    
    # 95%
    lower_95 = datos_ordenados[int(0.025 * n)]
    upper_95 = datos_ordenados[int(0.975 * n) - 1]
    
    # 99%
    lower_99 = datos_ordenados[int(0.005 * n)]
    upper_99 = datos_ordenados[int(0.995 * n) - 1]
    
    print(f"ICE (95%) para Santa Ana mes {mes}: [{lower_95}, {upper_95}]")
    print(f"ICE (99%) para Santa Ana mes {mes}: [{lower_99}, {upper_99}]")
```

```python
for mes in meses_floresta:
    datos = la_floresta_df[la_floresta_df['Mes'] == mes]['Ventas']
    x = np.sort(datos)
    y = np.arange(1, len(x)+1) / len(x)

    # Intervalos de Confianza Empíricos
    n = len(datos)
    datos_ordenados = np.sort(datos)
        
    # Intervalos con np.percentile
    lower_95 = np.percentile(datos, 2.5)
    upper_95 = np.percentile(datos, 97.5)

    lower_99 = np.percentile(datos, 0.5)
    upper_99 = np.percentile(datos, 99.5)

        
    print(f"ICE (95%) para La Floresta mes {mes}: [{lower_95}, {upper_95}]")
    print(f"ICE (99%) para La Floresta mes {mes}: [{lower_99}, {upper_99}]")
```

```python
santa_ana_df['DiaSemana'] = santa_ana_df['Fecha'].dt.day_name()
la_floresta_df['DiaSemana'] = la_floresta_df['Fecha'].dt.day_name()
```

```python
from statsmodels.distributions.empirical_distribution import ECDF
import matplotlib.pyplot as plt

dias = santa_ana_df['DiaSemana'].unique()

for dia in dias:
    ventas_sa = santa_ana_df[santa_ana_df['DiaSemana'] == dia]['Ventas']
    ventas_lf = la_floresta_df[la_floresta_df['DiaSemana'] == dia]['Ventas']

    ecdf_sa = ECDF(ventas_sa)
    ecdf_lf = ECDF(ventas_lf)

    plt.figure()
    plt.plot(ecdf_sa.x, ecdf_sa.y, label='Santa Ana')
    plt.plot(ecdf_lf.x, ecdf_lf.y, label='La Floresta')
    plt.title(f"ECDF - {dia}")
    plt.xlabel("Ventas")
    plt.ylabel("F(x)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
for dia in dias:
    plt.figure()
    sns.kdeplot(data=santa_ana_df[santa_ana_df['DiaSemana'] == dia], x='Ventas', label='Santa Ana', fill=True)
    sns.kdeplot(data=la_floresta_df[la_floresta_df['DiaSemana'] == dia], x='Ventas', label='La Floresta', fill=True)
    plt.title(f"Estimacion de Densidad - {dia}")
    plt.xlabel("Ventas")
    plt.ylabel("Densidad")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
niveles = [0.95, 0.99]

for dia in dias:
    print(f"\nIntervalos de Confianza para el dia {dia}:")

    # Datos de ventas por tienda
    ventas_sa = santa_ana_df[santa_ana_df['DiaSemana'] == dia]['Ventas']
    ventas_lf = la_floresta_df[la_floresta_df['DiaSemana'] == dia]['Ventas']

    # Calcular intervalos para Santa Ana
    for nivel in niveles:
        alpha = 1 - nivel
        lower = np.percentile(ventas_sa, 100 * alpha / 2)
        upper = np.percentile(ventas_sa, 100 * (1 - alpha / 2))
        print(f"Santa Ana - {int(nivel*100)}%: ({lower:.2f}, {upper:.2f})")

    # Calcular intervalos para La Floresta
    for nivel in niveles:
        alpha = 1 - nivel
        lower = np.percentile(ventas_lf, 100 * alpha / 2)
        upper = np.percentile(ventas_lf, 100 * (1 - alpha / 2))
        print(f"La Floresta - {int(nivel*100)}%: ({lower:.2f}, {upper:.2f})")
```

```python
# Calcular medias y desviaciones estándar
media_sa = santa_ana_df['Ventas'].mean()
media_lf = la_floresta_df['Ventas'].mean()
std_sa = santa_ana_df['Ventas'].std()
std_lf = la_floresta_df['Ventas'].std()

print("📊 Estadísticas generales:")
print(f"Santa Ana - Media: ${media_sa:.2f} | Desviación estándar: ${std_sa:.2f}")
print(f"La Floresta - Media: ${media_lf:.2f} | Desviación estándar: ${std_lf:.2f}")

# Comparación simple
if media_lf < media_sa:
    print("\n✅ La media de ventas de La Floresta es menor que la de Santa Ana.")
    diferencia = media_sa - media_lf
    print(f"La diferencia promedio es de ${diferencia:.2f} por día.")
else:
    print("\n❌ La media de ventas de La Floresta no es menor que la de Santa Ana.")
```

# Demostracion de normalidad para aplicar prueba de hipótesis

```python
def plot_sfo_traffic_analysis(data, numerical_columns, tienda):

    plt.figure(figsize=(4,12))

    for i, column in enumerate(numerical_columns):
        plt.subplot(4, len(numerical_columns), i + 1)
        sns.histplot(data[column], kde=True, bins=30)
        plt.title(f'Histogram of {column} ({tienda})')
        plt.xlabel(column)
        plt.ylabel('Frequency')

        plt.subplot(4, len(numerical_columns), i + 1 + len(numerical_columns))
        sns.boxplot(data[column])
        plt.title(f'Box-and-Whisker Plot of {column} ({tienda})')
        plt.xlabel(column)

        plt.subplot(4, len(numerical_columns), i + 1 + 2 * len(numerical_columns))
        sm.qqplot(data[column], line='45', fit=True, ax=plt.gca())
        plt.title(f'Q-Q Plot of {column} ({tienda})')


    plt.tight_layout()
    plt.show()
```

```python
tienda1 = 'Santa Ana'

plot_sfo_traffic_analysis(santa_ana_df, ["Ventas"], tienda1)
```

```python
tienda2 = 'La Floresta'

plot_sfo_traffic_analysis(la_floresta_df, ["Ventas"], tienda2)
```

```python
# Calcular las medias
mean_A = np.mean(la_floresta_df['Ventas'])
mean_B = np.mean(santa_ana_df['Ventas'])

# Calcular las desviaciones estándar muestrales
std_A = np.std(la_floresta_df['Ventas'], ddof=1)
std_B = np.std(santa_ana_df['Ventas'], ddof=1)

# Tamaños de las muestras
n_A = len(la_floresta_df['Ventas'])
n_B = len(santa_ana_df['Ventas'])

# Grados de libertad
degrees_of_freedom = n_A + n_B - 2

# Desviación estándar agrupada
sp = np.sqrt(((n_A - 1) * std_A**2 + (n_B - 1) * std_B**2) / degrees_of_freedom)

# Estadístico t
t_statistic = (mean_A - mean_B) / (sp * np.sqrt(1/n_A + 1/n_B))

# Nivel de significancia 99%
alpha = 0.005

# Valor crítico t (nosotros usaríamos este sólo para referencia, aquí aún se usa una librería)
t_critical = t_dist.ppf(1 - alpha, degrees_of_freedom)

# Cálculo manual del p-valor usando la función de distribución acumulada
p_value = 1 - t_dist.cdf(t_statistic, df=degrees_of_freedom)

# Imprimir los resultados
print(f'Grados de libertad: {degrees_of_freedom}')
print(f'Desviación estándar de la muestra A: {std_A}')
print(f'Desviación estándar de la muestra B: {std_B}')
print(f'Estadístico t: {t_statistic}')
print(f'Valor crítico t para alfa = {alpha}: {t_critical}')
print(f'Valor p: {p_value}')
print(f'Significancia: {alpha}')


if p_value < alpha:
    print(f"Rechazamos la hipótesis nula: La tienda {tienda2} vende significativamente más que la tienda {tienda1}.")
else:
    print(f"No rechazamos la hipótesis nula: No hay evidencia suficiente para afirmar que la tienda {tienda1} vende mas que la tienda {tienda2}.")
```

```python
# Nivel de significancia 95%
alpha = 0.025

# Valor crítico t (nosotros usaríamos este sólo para referencia, aquí aún se usa una librería)
t_critical = t_dist.ppf(1 - alpha, degrees_of_freedom)

# Cálculo manual del p-valor usando la función de distribución acumulada
p_value = 1 - t_dist.cdf(t_statistic, df=degrees_of_freedom)

# Imprimir los resultados
print(f'Grados de libertad: {degrees_of_freedom}')
print(f'Desviación estándar de la muestra A: {std_A}')
print(f'Desviación estándar de la muestra B: {std_B}')
print(f'Estadístico t: {t_statistic}')
print(f'Valor crítico t para alfa = {alpha}: {t_critical}')
print(f'Valor p: {p_value}')
print(f'Significancia: {alpha}')


if p_value < alpha:
    print(f"Rechazamos la hipótesis nula: La tienda {tienda2} vende significativamente más que la tienda {tienda1}.")
else:
    print(f"No rechazamos la hipótesis nula: No hay evidencia suficiente para afirmar que la tienda {tienda1} vende mas que la tienda {tienda2}.")
```
